```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/internship/Data_Set_For_Task/3) Sentiment dataset.csv")

# Keep only needed columns
df = df[['Text', 'Sentiment']]

# Simplify labels
def simplify_sentiment(label):
    label = str(label).lower()

    if 'pos' in label or 'joy' in label or 'happy' in label:
        return 'Positive'
    elif 'neg' in label or 'sad' in label or 'hate' in label:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Sentiment'].apply(simplify_sentiment)

# Features & target
X = df['Text']
y = df['Sentiment']

# Convert text
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
